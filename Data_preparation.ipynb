```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
data = pd.read_csv('/content/drive/MyDrive/TSF Intern/Task 4/globalterrorismdb_0718dist.csv' , encoding="ISO-8859-1")
df = data.copy()
df.head(10)
```

```python
df.describe()
```

```python
#Dropping all the un-needed columns including all the null values
#note this will drop also some important columns so I'll return them back later and deal with all null values by replacing them.
#df = df.dropna()
df = df.dropna(axis=1)
```

```python
pd.set_option('display.max_columns', None)
df.head(10)
```

```python
#Dropping more un-needed column
df = df.drop(['vicinity'], axis = 1)
```

```python
#Dropping columns with inclusion criterias
df = df.drop(['crit1', 'crit2', 'crit3'], axis = 1)
```

```python
df.isnull().sum()
```

```python
#get killsn, get nwounds, int_any(International Attack) , get latitude , get longitude , remove property,propextent and get	propextent_txt(proprty damage)
#terrorism_df['City']=terrorism_df['City'].fillna('Unknown')
#nba["College"].fillna("No College", inplace = True)
```

```python
#Dropping the countrey & region counts as we wants their names only
#renaming the countrey_txt & region_txt
df = df.drop(['country', 'region'], axis = 1)
df.rename(columns = {'country_txt':'country'}, inplace = True)
df.rename(columns = {'region_txt':'region'}, inplace = True)
```

```python
#Returning the city & provstate columns
df = df.join(data['city'])
df = df.join(data['provstate'])
```

```python
#Letting the INT_ANY column and dropping all other columns that provides same data
#Renaming INT_ANY
df = df.drop(['INT_LOG', 'INT_IDEO', 'INT_MISC'], axis = 1)
df.rename(columns = {'INT_ANY':'International Attack'}, inplace = True)
```

```python
#Returning the number of kills and nkills in US
#replacing the null values with 0
df = df.join(data['nkill'])
df = df.join(data['nkillus'])
df['nkill'].fillna(0 , inplace =True)
df['nkillus'].fillna(0 , inplace =True)
```

```python
#Returning the number of wounds and nwounds in US
#replacing the null values with 0
df = df.join(data['nwound'])
df = df.join(data['nwoundus'])
df['nwound'].fillna(0 , inplace =True)
df['nwoundus'].fillna(0 , inplace =True)
```

```python
#Dropping the property as it is covered in propextent_txt so we will also retrieve propextent_txt
#replacing the null values in propextent_txt with unknown
#renaming propextent_txt
df = df.drop(['property'], axis = 1)
df = df.join(data['propextent_txt'])
df['propextent_txt'].fillna('Unknown' , inplace =True)
df.rename(columns = {'propextent_txt':'Property Damage'}, inplace = True)
```

```python
#Retrieving latitude & longitude
df = df.join(data['latitude'])
df = df.join(data['longitude'])
```

```python
#Renaming some columns in order to be understandable
df.rename(columns = {'attacktype1':'Attack Num'}, inplace = True)
df.rename(columns = {'attacktype1_txt':'Attack Type'}, inplace = True)
df.rename(columns = {'targtype1':'Traget Num'}, inplace = True)
df.rename(columns = {'targtype1_txt':'Target Type'}, inplace = True)
df.rename(columns = {'weaptype1':'Weapon Num'}, inplace = True)
df.rename(columns = {'weaptype1_txt':'Weapon Type'}, inplace = True)
```

```python
df.isnull().sum()
```

```python
#Dropping all the rows left with null values
df_cleaned = df.dropna()
```
